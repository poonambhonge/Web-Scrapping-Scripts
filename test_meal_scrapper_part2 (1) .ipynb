```python
#List of calories requied.
calorie=["12000"]
#Add meal split here.
m_split="3 meals"
```

```python
#install chrome driver

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
sno=0

for turn in range(0,len(calorie)):
    for x in range(20):
#         try:
            #launch website
            driver.get("https://www.eatthismuch.com/")

            #getting generate button element
            generate_button = WebDriverWait(driver, 60).until(EC.element_to_be_clickable((By.CLASS_NAME, 'gen_button')))

            #getting elements with class nav_items
            nav_items = WebDriverWait(driver, 60).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "nav-item")))

            #extracting diet items from above elements
            diet_items = [nav_items[4], nav_items[5], nav_items[6], nav_items[7], nav_items[8], nav_items[9]]

            #particular diet category (e.g. Vegetarian)
            diet_category = diet_items[1].text

            diet_items[1].click()           #clicking that diet category

            #getting calories input element
            input_element = WebDriverWait(driver, 60).until(EC.presence_of_element_located((By.ID, "cal_input")))

            # sending calories in input
            input_element.clear()
            input_element.send_keys(calorie[turn])

            #getting number of meals input element
            num_meals_input = WebDriverWait(driver, 60).until(EC.presence_of_element_located((By.ID,'num_meals_selector')))
            num_meals_input.send_keys(m_split)

            generate_button.click()

            # cancelling popover
            cancel_popover_element = WebDriverWait(driver, 60).until(EC.element_to_be_clickable((By.CLASS_NAME, 'cancel_popover')))
            cancel_popover_element.click()
            sleep(1)

            link=dict()

            recipe_elements = WebDriverWait(driver, 65).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "print_name")))

            for j in range(0,int(len(recipe_elements))):
                sleep(1)
                if recipe_elements[j].text== '':
                    break

                sleep(1)
                recipe_elements[j].click()
                sleep(1)
                recipe_link = WebDriverWait(driver, 60).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "recipe_permalink")))
                link[recipe_elements[j].text]=recipe_link[0].get_attribute('value')
                wait = WebDriverWait(driver, 60)
                recipe_close = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="food_object_modal"]/div/div/div[1]/button')))
                print(recipe_close)
                recipe_close.click()


            sleep(1)
            # Meal Header: Breakfast, Lunch, Dinner, Snack
            meal_header_elements = WebDriverWait(driver, 60).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "meal_header")))
            meals=[]
            cal=[]
            for j in range(len(meal_header_elements)):
                if meal_header_elements[j].text== '':
                    break
                meals.append((meal_header_elements[j].text.split('\n'))[0])

            recipe_elements = WebDriverWait(driver, 60).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "print_name")))
            recipe_names = []
            for j in range(0,int(len(recipe_elements))):
                if recipe_elements[j].text== '':
                    break
                recipe_names.append(recipe_elements[j].text)

            meal_content_elements = WebDriverWait(driver, 60).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "meal_content")))
            content = []
            for j in range(0,int(len(meal_content_elements))):
                if meal_content_elements[j].text == '':
                    break
                content.append(meal_content_elements[j].text)

            coll = dict()

            for j in range(len(content)):
                coll[j]=list()
            for j in recipe_names:
                for k in range(len(content)):
                    if content[k].__contains__(j):
                        coll[k].append(link[j])
            for j in range(len(content)):
                coll[j]=" & ".join(coll[j])

            recipe_dict=dict()

            for j in meals:
                recipe_dict[j]=list()
            for k in range(len(meals)):
                recipe_dict[meals[k]].append(coll[k])
            for k in set(meals):
                recipe_dict[k]=" | ".join(recipe_dict[k])


            # creating dataframe
            df.loc[sno,"S.No"]=sno+1
            df.loc[sno,"Total Calories"]=calorie[turn]
            df.loc[sno,"Meal Split"]=m_split
            uui=str(uuid.uuid4())
            df.loc[sno,"Iteration id"]=uui
            for j in set(meals):
                df.loc[sno,j]=recipe_dict[j]
            df.loc[sno,"Category"]=diet_category
            sno=sno+1

                #regenerate button
    #             regen_element = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "day_refresh_button")))
    #             regen_element[0].click()
    #             sleep(5)

#         except TimeoutException:
#             print("Time_out at ")

#         except ElementClickInterceptedException:
#             print("Element intercepted")
```
